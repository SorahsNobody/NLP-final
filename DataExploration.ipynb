# Data exploration

## Load libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load data

```python
data = pd.read_csv('all_feat_final_new.csv')
data.drop(columns = ["Unnamed: 0"], inplace = True, axis = 1)
data.rename(columns = {'clicked':'class'}, inplace = True)
data
```

## Data summary

```python
data.describe()
```

```python
class_counts = data['class'].value_counts()
class_counts
```

```python
class_name = {0: "Not clicked", 1: "Clicked"}
class_counts.rename(class_name, inplace=True)

class_counts.plot.pie(autopct='%1.1f%%', labels=class_counts.index)
plt.axis('equal')
plt.title('Distribution of Classes', fontsize=20)
plt.show()
```

## Into features (feature imp)

```python
d = data['at_Chara_C'].describe()
d
```

```python
# top ten maximum values
top_10 = data['at_Chara_C'].nlargest(10)
top_10_res = data.loc[data['at_Chara_C'].isin(top_10), 'class']
top_10_res.sort_values(ascending=False)
```

```python
sns.barplot(x='class', y='at_Chara_C', data=data)

plt.title('Barplot of at_Chara_C by class (1st)')
plt.xlabel('Class')
plt.ylabel('at_Chara_C')
plt.show()
```

```python
sns.histplot(data=data, x="at_Chara_C", bins=200, kde = True, line_kws={'color': 'black', 'lw': 2})
```

```python
# 2nd
sns.barplot(x='class', y='to_FuncW_C', data=data)

plt.title('Barplot of to_FuncW_C by class (2nd)')
plt.xlabel('Class')
plt.ylabel('to_FuncW_C')
plt.show()
```

```python
sns.barplot(x='class', y='at_Sylla_C', data=data)

plt.title('Barplot of at_Sylla_C by class (2nd)')
plt.xlabel('Class')
plt.ylabel('at_Sylla_C')
plt.show()
```
